Introduction:
**Analyzing employee attrition prediction is crucial for several reasons:**

**Cost Reduction:** Employee turnover can be expensive for businesses due to recruitment, training, and onboarding costs. By predicting attrition, businesses can proactively take measures to retain valuable employees, thus reducing turnover costs.

**Workforce Planning:** Anticipating attrition allows businesses to plan their workforce more effectively. They can identify areas with high turnover rates and take proactive steps such as succession planning, talent development, or hiring to mitigate the impact of potential departures.

**Maintaining Productivity:** Employee turnover can disrupt workflow and productivity, especially if key positions remain vacant for a long time. Predictive analysis helps in identifying potential departures in advance, allowing businesses to make necessary arrangements to minimize disruptions and maintain productivity levels.

**Employee Satisfaction and Engagement:** Analyzing attrition prediction can provide insights into employee satisfaction and engagement levels within the organization. By identifying factors contributing to attrition, such as dissatisfaction with work environment, lack of growth opportunities, or poor management, businesses can take corrective actions to improve employee morale and retention.

**Competitive Advantage:** High employee turnover can negatively impact a company's reputation and competitiveness in the market. Businesses that effectively manage attrition through predictive analysis are more likely to retain top talent, maintain continuity in operations, and build a positive employer brand, giving them a competitive edge.

Overall, analyzing employee attrition prediction enables business owners to make informed decisions, reduce costs, enhance workforce planning, maintain productivity, improve employee satisfaction, and gain a competitive advantage in the market.

# TASK #6: UNDERSTAND THE INTUITION BEHIND LOGISTIC REGRESSION CLASSIFIERS

# TASK #7: TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
# Testing Set Performance
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

**MINI CHALLENGE #5:**
- **Train an XGboost algorithm instead of logistic regression**
- **Compare XGBoost model performance to Logistic Regression**

```python
from xgboost import XGBClassifier

model = XGBClassifier(use_label_encoder=False)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```
